```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('clusters.csv')
```

```python
n_clusters = (2, 3, 4)
for name,d in data.groupby('name'):
    for n in n_clusters:
        print(name, ': k =', n)
        labels = KMeans(n_clusters=n).fit_predict(d[['x', 'y']].values)
        fig = plt.subplots()
        sns.scatterplot(data=d.assign(label=labels), x='x', y='y', hue='label', legend=False)
        plt.show()
```
